```python
import json
def read_jsonl(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data.append(json.loads(line.strip()))
    return data
```

```python
import re
def process_answer(string, type='not_quantized'):
    pattern = r'<[^>]*[A-Z][^>]*>'    
    # Use findall to extract all matching substrings
    matches = re.findall(pattern, string)

    try:
        match = matches[0]
        ans = match.split(': ')[1]
        ans = ans.split('>')[0]
        ans=ans.strip()
        if len(ans) == 1:
            return ans
        else: print(ans)
    except:
        if type == 'quantized':
            a=1
        #print(string)
        return 'A'
    return matches
```

```python
acc = 0
for i, (question, answer) in model_base.iterrows():
    q = question.split('### Question: ')[1].split('### Answer:')[0][:-12]
    q = 'Question: ' + q
    ground_truth_answer = ground_truth[ground_truth['question'] == q]['answer'].values
    pred_base_model = answer

    pred = (process_answer(pred_base_model[0]))
    pred_quantized_model = model_quantized.iloc[i].policy
    #(process_answer(pred_quantized_model[0], 'quantized'))
    if pred == ground_truth_answer:
        acc += 1
```
